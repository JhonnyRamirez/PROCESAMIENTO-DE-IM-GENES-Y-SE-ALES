```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# 1. Cargar la imagen 
img_original = Image.open("foto_capturada1.jpg")

# 2. Procesamiento para detectar lentes y silueta
img_gris = img_original.convert("L")

# Aplicamos el filtro de deteccin de bordes
img_bordes = img_gris.filter(ImageFilter.FIND_EDGES)

# 3. Visualizacin lado a lado
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

# Izquierda: Original
axs[0].imshow(img_original)
axs[0].set_title("1. Foto Original")
axs[0].axis('off')

# Derecha: Bordes (Lentes y Silueta)
axs[1].imshow(img_bordes, cmap='gray')
axs[1].set_title("2. Deteccin de Bordes (PIL)")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```
