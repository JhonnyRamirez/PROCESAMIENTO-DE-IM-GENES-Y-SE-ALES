```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# 1. Cargar la imagen 
img_original = Image.open("foto_micara.jpg")

# --- OPERACIONES GEOMTRICAS Y FILTROS ---

# A. Rotacin de 180 grados
img_rotada = img_original.rotate(180)

# B. Reflejo 
img_reflejada = img_original.transpose(Image.FLIP_LEFT_RIGHT)

# C. Enfoque Mejorado
img_enfocada = img_original.filter(ImageFilter.SHARPEN)

# --- VISUALIZACIN ---
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

axs[0].imshow(img_original)
axs[0].set_title("Original")

axs[1].imshow(img_rotada)
axs[1].set_title("Rotada 180")

axs[2].imshow(img_reflejada)
axs[2].set_title("Reflejo (Espejo)")

axs[3].imshow(img_enfocada)
axs[3].set_title("Enfoque Mejorado")

# Quitar ejes para que se vea limpio
for ax in axs: ax.axis('off')

plt.tight_layout()
plt.show()
```
